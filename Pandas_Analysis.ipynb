# Q4 2024 Sales Data Cleaning & Analysis
### Pandas Data Analysis Project
This notebook focuses on cleaning a messy Q4 2024 e-commerce dataset and generating meaningful business insights using Pandas.

## Task 1: Load and Inspect the Data

In this step:
- We import required libraries (Pandas & NumPy)
- Create the DataFrame using the provided dataset
- Inspect structure and missing values

Understanding the dataset structure is essential before performing any cleaning operations.

```python
import pandas as pd
import numpy as np

data = {
    'transaction_id': range(1, 21),
    'date': pd.date_range('2024-10-01', periods=20, freq='D'),
    'region': ['North', 'South', 'East', 'West', None, 'North', 'South', None, 'East', 'West',
               'North', 'South', 'East', 'West', 'North', None, 'East', 'West', 'North', 'South'],
    'product_category': ['Electronics', 'Clothing', None, 'Books', 'Electronics', 'Home',
                         'Clothing', 'Books', 'Electronics', None, 'Home', 'Clothing',
                         'Books', 'Electronics', 'Home', 'Clothing', 'Books', 'Electronics',
                         'Home', 'Clothing'],
    'sales_amount': [1200, 450, 890, None, 1500, 670, None, 340, 2100, 780,
                     560, None, 420, 1800, 920, 510, 380, None, 1100, 640],
    'quantity': [2, 5, 3, 1, None, 4, 2, 3, 1, 5,
                 3, 2, None, 1, 4, 3, 2, 1, None, 4],
    'customer_age': [25, 34, None, 45, 29, None, 38, 52, 27, 41,
                     33, None, 48, 26, 35, 42, None, 31, 39, 44],
    'payment_method': ['Credit Card', 'UPI', 'Cash', 'Debit Card', 'Credit Card',
                       'UPI', 'Cash', None, 'Credit Card', 'UPI', 'Debit Card',
                       'Cash', 'Credit Card', None, 'UPI', 'Cash', 'Debit Card',
                       'Credit Card', 'UPI', None]
}

df = pd.DataFrame(data)
```

### Inspect Data Structure

We use `df.info()` to:
- Check column data types
- View non-null counts
- Understand missing values at a high level

```python
df.info()
```

### Check Missing Values

We use `df.isna().sum()` to:
- Count missing values in each column
- Identify which columns require cleaning

```python
df.isna().sum()
```

## Task 2: Handle Missing Values

Now we clean the dataset using appropriate strategies based on data type:

- Categorical columns → Fill with Mode
- Numerical columns → Fill with Mean or Median
- Time/Sequential data → Forward Fill
- Drop rows when necessary (payment_method)

This ensures the dataset is ready for analysis.

### Fill Missing Values in Categorical Columns

We fill missing values in:
- `region`
- `product_category`

Using the **mode (most frequent value)** because it preserves category distribution.

```python
df['region'].fillna(df['region'].mode()[0],inplace=True)
df['product_category'].fillna(df['product_category'].mode()[0],inplace=True)
```

### Fill Missing Sales Amount

We replace missing values in `sales_amount` using the **median**.

Median is chosen because:
- It is robust to outliers
- It represents central tendency better for skewed sales data

```python
df['sales_amount'].fillna(df['sales_amount'].median(),inplace=True)
```

### Fill Missing Quantity using Forward Fill

Since transaction data is sequential,
we use **forward fill (ffill)** to propagate the last valid quantity value.

```python
df['quantity'].fillna(method='ffill',inplace=True)
```

### Fill Missing Customer Age

We replace missing values in `customer_age` using the **mean age**.
The result is rounded to maintain integer values.

```python
df['customer_age'].fillna(round(df['customer_age'].mean()),inplace=True)
df['customer_age']=df['customer_age'].astype(int)
```

### Drop Rows with Missing Payment Method

Payment method is critical for transaction analysis.
Instead of guessing values, we remove rows where `payment_method` is missing.

```python
df.dropna(subset=['payment_method'],inplace=True)
```

### Verify Data Cleaning

We confirm that no missing values remain in the dataset.
This ensures the dataset is clean and analysis-ready.

```python
df.isna().sum()
```

## Task 3: GroupBy Analysis

Now that the dataset is clean, we perform aggregation analysis to generate business insights.

We analyze:
- Total sales by region
- Average sales by product category
- Region-product level performance

### Total Sales by Region

We group by `region` and calculate total sales.
This helps identify the highest revenue-generating region.

```python
Total_Sales_Region=df.groupby('region')['sales_amount'].sum()
print(Total_Sales_Region)
```

### Average Sales by Product Category

We calculate the average sales for each product category
to understand category-level performance.

```python
Average_Sales_Category=df.groupby('product_category')['sales_amount'].mean()
print(Average_Sales_Category)
```

### Sales & Quantity by Region and Product Category

We perform multi-column grouping:
- region
- product_category

Then calculate:
- Total sales
- Total quantity

We use `reset_index()` to convert the hierarchical index into a flat table.

```python
Grouped=df.groupby(['region','product_category'])[['sales_amount','quantity']].sum()
Grouped=Grouped.reset_index()
print(Grouped)
```

### Top 3 Region-Product Combinations

We sort the grouped results by sales_amount
and display the top 3 performing combinations.

```python
Top_3=Grouped.sort_values('sales_amount',ascending=False).head(3)
print(Top_3)
```

## Task 4: Custom Aggregation

We create a custom function to calculate the sales range
(Maximum - Minimum sales).

This helps measure sales variability across regions.

### Sales Range by Region

We apply the custom function using `.agg()`
to calculate sales variation for each region.

```python
def sales_range(x):
  return x.max() - x.min()
```

### Multiple Aggregations by Region

Using `.agg()` we calculate multiple metrics at once:

For `sales_amount`:
- Sum
- Mean
- Maximum

For `quantity`:
- Sum
- Minimum

This provides a complete performance summary per region.

```python
Range_By_Region=df.groupby('region')['sales_amount'].agg(sales_range)
print(Range_By_Region)
```

## Final Insights & Conclusion

- Identified top-performing regions based on total sales.
- Electronics category shows strong revenue performance.
- Certain region-product combinations significantly outperform others.
- Sales variation differs across regions.
- The dataset is now fully cleaned and ready for advanced analysis or visualization.

This structured cleaning and aggregation ensures reliable business insights.

```python
Multiple_Metrics=df.groupby('region').agg({
    'sales_amount':['sum','mean','max'],
    'quantity':['sum','min']
})
print(Multiple_Metrics)
```
